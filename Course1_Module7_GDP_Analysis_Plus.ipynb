```python
# User defined function to facilitate plotting

def bar_plot(df, ycol, xticks, yscale, title):
    
    # set figure size for larger figure
    plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')

    # bar plot
    ax = sns.barplot(x=df.index, y=ycol, data=df)

    if xticks==True:
        
        # rotate and align the tick labels to look better
        ax.set_xticklabels(ax.get_xticklabels(), rotation=50, ha='right')

    if yscale==True:
        
        # changing to a logarithmic scale so that the range can fit into the size of the plot.
        plt.yscale('log')

    # set title
    ax.set_title(title)
    
    # adjust plot parameters so that the plot fits in to the figure area
    plt.tight_layout()

    # show plot
    return plt.show()
```

```python
bar_plot(df1a_mean, 'Average_Growth', True, False, 'Average Growth in GSDP of states for 2013-14, 2014-15 and 2015-16')
```

```python
# User defined function to facilitate plotting

def reg_plot(df, x, y, title):

    # set figure size for larger figure
    plt.figure(num=None, figsize=(9, 6), dpi=69, facecolor='w', edgecolor='k')

    ax = sns.regplot(x=x, y=y, data=df, color="g")

    # set title
    ax.set_title(title)

    # adjust plot parameters so that the plot fits in to the figure area
    plt.tight_layout()

    # show plot
    return plt.show()
```
